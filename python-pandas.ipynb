```python
from sklearn import datasets

# decoded species name to use as str values
iris = datasets.load_iris(as_frame=True)
data = iris["data"]
decoded_targets = [iris["target_names"][idx] for idx in iris["target"]]
data["species"] = decoded_targets
data
```

```python
# filter methods
data.loc[data["species"] == "virginica"].head()
```

```python
query_list = {"virginica", "setosa"}
data.loc[data["species"].isin(query_list)]
```
